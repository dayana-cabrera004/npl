```python
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import Layout, Box, HTML

# Extended dataset information with multiple diseases
dataset_info = {
    "covid-19": [
        {
            "name": "COVID-19 Global Data",
            "url": "https://github.com/datasets/covid-19"
        },
        {
            "name": "COVID-19 Data Repository by Johns Hopkins University",
            "url": "https://github.com/CSSEGISandData/COVID-19"
        },
        {
            "name": "COVID-19 Open Data by Google Cloud",
            "url": "https://github.com/GoogleCloudPlatform/covid-19-open-data"
        }
    ],
    "malaria": [
        {
            "name": "Malaria Data from WHO",
            "url": "https://www.who.int/malaria/data"
        },
        {
            "name": "Malaria Incidence Data",
            "url": "https://data.worldbank.org/indicator/SH.MLR.TERT"
        }
    ],
    "ebola": [
        {
            "name": "Ebola Data Repository",
            "url": "https://github.com/opendata-berkeley/ebola"
        },
        {
            "name": "Ebola Outbreak Data",
            "url": "https://www.kaggle.com/datasets/patrickb1912/ebola-outbreak"
        }
    ],
    "tuberculosis": [
        {
            "name": "Global Tuberculosis Report by WHO",
            "url": "https://www.who.int/tb/publications/global_report"
        }
    ],
    "dengue": [
        {
            "name": "Dengue Fever Data",
            "url": "https://www.kaggle.com/datasets/aaraki/dengue-fever-data"
        }
    ],
    "heart disease": [
        {
            "name": "Heart Disease Data (Cleveland Dataset)",
            "url": "https://archive.ics.uci.edu/ml/datasets/heart+disease"
        }
    ],
    "diabetes": [
        {
            "name": "Diabetes dataset from Kaggle",
            "url": "https://www.kaggle.com/datasets"
        },
        {
            "name": "Diabetes data from NIDDK",
            "url": "https://www.niddk.nih.gov"
        }
    ],
    "cancer": [
        {
            "name": "Breast Cancer dataset from UCI",
            "url": "https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Original)"
        },
        {
            "name": "Cancer Data from NCI",
            "url": "https://www.cancer.gov"
        }
    ],
    "cholera": [
        {
            "name": "Cholera Data from WHO",
            "url": "https://www.who.int/news-room/fact-sheets/detail/cholera"
        }
    ],
    "inflammatory breast cancer": [
        {
            "name": "Inflammatory Breast Cancer Data from NCI",
            "url": "https://www.cancer.gov"
        },
        {
            "name": "Inflammatory Breast Cancer Data from Kaggle",
            "url": "https://www.kaggle.com/datasets"
        },
        {
            "name": "Inflammatory Breast Cancer Dataset by the Breast Cancer Surveillance Consortium",
            "url": "https://www.cancer.gov/about-cancer/types/breast"
        }
    ]
}

# Function to search datasets based on disease
def search_disease(disease_name):
    disease_name = disease_name.lower().strip()

    # Check if the disease exists in the dataset_info dictionary
    if disease_name in dataset_info:
        datasets = dataset_info[disease_name]
        result_text = f"Found {len(datasets)} dataset(s) for '{disease_name}':\n\n"
        for dataset in datasets:
            result_text += f"- {dataset['name']}: {dataset['url']}\n"
    else:
        result_text = f"No datasets found for '{disease_name}'. Please check your spelling or try another disease."

    # Update output with result
    output_label.value = result_text

# Function to display available diseases
def show_available_diseases():
    available_diseases = "\n".join(dataset_info.keys())
    return f"<b>Available diseases to search for:</b>\n\n{available_diseases}\n\nPlease type a disease name to get related datasets."

# Create a more professional UI with colors and fonts
header_style = 'font-size: 20px; font-weight: bold; color: #2a8b8d; margin-bottom: 20px;'
instruction_style = 'font-size: 16px; color: #4a4a4a; margin-bottom: 15px;'
button_style = 'background-color: #4CAF50; color: white; font-size: 16px; padding: 10px 24px; border-radius: 5px;'
output_style = 'font-size: 16px; color: #333; background-color: #f9f9f9; padding: 10px; border-radius: 5px; border: 1px solid #ddd;'

# Disease info to display
diseases_info = show_available_diseases()

# Create UI widgets
diseases_label = widgets.HTML(value=f"<div style='{header_style}'>Search for Public Healthcare Datasets</div>")
instructions_label = widgets.HTML(value=f"<div style='{instruction_style}'>Type a disease name (e.g., 'covid-19') and click the button to find datasets.</div>")
disease_input = widgets.Text(placeholder='Type a disease name here...', layout=Layout(width='100%'))
search_button = widgets.Button(description="Search Datasets", style={'button_color': '#4CAF50'}, layout=Layout(width='100%'))
output_label = widgets.HTML(value="", layout=Layout(width='100%', min_height='150px'))

# Function to handle the button click event
def on_button_click(b):
    disease_name = disease_input.value
    if disease_name:
        search_disease(disease_name)
    else:
        output_label.value = "<b style='color:red;'>Please enter a disease name.</b>"

# Attach the button click event to the function
search_button.on_click(on_button_click)

# Display everything
display(diseases_label, instructions_label, disease_input, search_button, output_label)
```
